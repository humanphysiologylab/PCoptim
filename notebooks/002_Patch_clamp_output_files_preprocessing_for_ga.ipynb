### How it works

If you want to read file.atf (or save it in the format appropriate to ga) - use function
>*trace_from_atf*(path, save, new_path)

where :

&emsp; *path*  - is path to file.atf

&emsp;  *save* - True/False, if you want to save your trace in file

&emsp;  *new_path* - path there you want to save your trace(default path to trace is PCoptim/data/traces)


If you want to read file.abf use 
>*trace_from_abf*(path, save, new_path).

You can also get the protocol:
>*protocol_from_abf*(path, save, new_path)

and potential between sweeps:
>*initial_protocol_from_abf*(path, save, new_path)

for ga optimization from the file.abf.

By default imported data is saved in subdirectories *protocols, initial_protocols* and *traces* of PCoptim/data directory. The setting 'save' parameter to False imports the data without saving the CSV.

```python
atf_path = '../data/raw_data/test_atf.atf'
atf = trace_from_atf(atf_path, save=False)
plt.plot(atf)
```
